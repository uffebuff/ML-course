```python
# Import packages
import feedparser

################################ RSS FEED Parser ################################

RSS_URLS = ["https://www.dn.se/rss/"]  # Lägg till fler URL:er om det behövs

# Create an empty list called ‘posts’ here
posts = []

# Loop through the RSS feeds and extract the articles
for url in RSS_URLS:
    feed = feedparser.parse(url)  # Läs ner RSS-flödet
    for entry in feed.entries:  # Iterera genom varje artikel i flödet
        posts.append({
            "title": entry.title,       # Artikelns titel
            "link": entry.link,         # Artikelns länk
            "published": entry.published,  # Publiceringsdatum (om tillgängligt)
        })

# Print the results
print(posts)
```
